```python
import requests
from bs4 import BeautifulSoup
from lxml import html
```

```python
headers={'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/111.0.0.0 Safari/537.36',
         "Accept-Language": "en-US,en;q=0.5"}
url = "https://www.imdb.com/title/tt0249603"
page = requests.get(url, headers=headers)
```

```python
with open("output.html", "w") as file:
    file.write(page.text)
```

```python
tree = html.fromstring(page.text)
```

```python
title = tree.xpath("//span[@data-testid='hero__primary-text']/text()")[0]
title
```

```python
director = tree.xpath("//a[@class='ipc-metadata-list-item__list-content-item ipc-metadata-list-item__list-content-item--link']/text()")[0]
director
```

```python
brief_plot = tree.xpath("//div[@class='sc-f5ef05d0-0 gutvDK']/descendant::div[@class='ipc-html-content-inner-div']/text()")
brief_plot
```

```python
genres = tree.xpath("//li[@data-testid='storyline-genres']/text()")
genres
```
